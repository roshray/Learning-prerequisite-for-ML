```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
    movies['title'].sort_values()
```

```python
movies.sort_values('title')
```

```python
# for multiple title sorting use list
movies.sort_values(['content_rating','duration'])
```

```python
#filter rows in DataFrame
```

```python
movies.head()
```

```python
booleans = []
for length in movies.duration:
    if length >=200:
        booleans.append(True)
    else:
        booleans.append(False)
        
```

```python
booleans[0:5]
```

```python
series = pd.Series(booleans)
```

```python
series.head()
```

```python
movies[series]
```

```python
#shortcut of above
#Display the same out put
movies[movies.duration >= 200]
```

```python
#only genre
movies[movies.duration >= 200].genre
```

```python
# multiple filter in pandas  dataframe

movies[(movies.duration >= 200) & (movies.genre=='Drama')]
```

```python
#Both produces same output.

#movies[(movies.genre =='Action') | (movies.genre == 'Drama') | (movies.genre == 'Crime')]
movies[movies.genre.isin(['Action','Drama','Crime'])]
```
